```python
import os
import requests
from splinter import Browser
from bs4 import BeautifulSoup
import pymongo
import pandas as pd
```

```python
executable_path = {'executable_path':'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
title = soup.find('div', class_='content_title')
print(f'News Title' + ' = ' + title.text)

paragraphs = soup.find('div', class_='article_teaser_body')
print(f'News Paragraph' + ' = ' + paragraphs.text)
```

```python
photo_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(photo_url)
```

```python
p_html = browser.html
soup_p = BeautifulSoup(p_html, 'html.parser')
```

```python
photos = soup_p.find('img', class_='thumb')['src']
pre_url = 'http://www.jpl.nasa.gov'
featured_image_url = pre_url + photos
print(featured_image_url)
```

```python
url_mars = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
url_mars_base = 'https://astrogeology.usgs.gov'

browser.visit(url_mars)

m_html = browser.html

soup_m = BeautifulSoup(m_html, 'html.parser')

list_names = []
list_links = []
```

```python
titles = soup_m.find_all('div', class_='description')
for title in titles:
    
    photo_name = title.find('h3').text
    photo_link = title.find('a')['href']
    
    list_names.append(photo_name)

    url_mars_enhanced = url_mars_base + photo_link
    
    browser.visit(url_mars_enhanced)
    m_e_html = browser.html
    
    soup_me = BeautifulSoup(m_e_html, 'html.parser')
    
    enhanced = soup_me.find('div', class_='downloads')
    
    for enhance in enhanced:
        
        the_link = enhanced.a['href']
        
        list_links.append(the_link)
        
        print
```
